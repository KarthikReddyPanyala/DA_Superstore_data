```python
import pandas as pd 
data=pd.read_csv('C://Users//karth//OneDrive//Desktop//Retail_sales_analytics//data//Sample - Superstore.csv',encoding='ISO-8859-1')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
#convert dates to datetimes
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.dtypes
```

```python
data['Order Year']=data['Order Date'].dt.year
data['Order Month']=data['Order Date'].dt.month_name()
data['Order Day']=data['Order Date'].dt.day
data['Order Weekday']=data['Order Date'].dt.day_name()
```

```python
data.info()
data.head()
```

```python
data.drop(['Row ID','Postal Code'],axis=1,inplace=True)
```

```python
data.info()
```

```python
#most profitable products
top_10=data.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.barh(top_10.index,top_10.values)
```

```python
import matplotlib.pyplot as plt
top10_regions=data.groupby('State')['Profit'].sum().sort_values()
plt.figure(figsize=(15,10))
plt.barh(top10_regions.index,top10_regions.values)
```

```python
monthly_sales = data.groupby('Order Month')['Sales'].sum().reindex(['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December'])
monthly_sales.plot(kind='bar', title='Monthly Sales Trends', ylabel='Sales ($)')
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,10))
sns.scatterplot(x='Discount',y='Profit',data=data,hue='Category')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.axhline(0,color='red',linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt 
loss_subcategories=data.groupby('Sub-Category')['Profit'].sum().sort_values().head(5)
plt.figure(figsize=(15,10))
plt.barh(loss_subcategories.index,loss_subcategories.values)
plt.show()
```

```python
top5_categories=data.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(5)
top5_categories.plot(kind='barh',title='Sorted Total Sales By Category',ylabel='Sales',figsize=(15,10))
plt.show()
```

```python
top5_subcategories=data.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(5)
top5_subcategories.plot(kind='barh',title='Sorted Total Sales By Sub-Category',ylabel='Sales',figsize=(15,10))
plt.show()
```

```python
monthly_sales=data.groupby(['Order Year','Order Month'])['Sales'].sum().reset_index()
monthly_sales['Month Year'] = monthly_sales['Order Month']+' '+monthly_sales['Order Year'].astype(str)
plt.figure(figsize=(20,15))
plt.plot(monthly_sales['Month Year'],monthly_sales['Sales'],marker='o')
plt.xticks(rotation=45)
```

```python
#weekday vs weekend analysis
data['Day Type']=data['Order Weekday'].apply(lambda x: 'Weekend' if x in ['Saturday','Sunday'] else 'Weekday')
day_type_sales=data.groupby('Day Type')['Sales'].sum().reset_index()
plt.figure(figsize=(15,10))
plt.bar(day_type_sales['Day Type'],day_type_sales['Sales'])
plt.title('Total sales :  weekday vs weeknd')
plt.show()
```

```python
#shipping delay plot
data['Shipping Delay']=(data['Ship Date']-data['Order Date']).dt.days
region_delay=data.groupby('Region')['Shipping Delay'].mean().reset_index()
plt.figure(figsize=(15,10))
plt.barh(region_delay['Region'],region_delay['Shipping Delay'])
plt.title('Aversage Shipping Delay by region')
plt.xlabel('Average  Delay')
plt.ylabel('Region')
plt.show()
```

```python
data['Shipping Delay']=(data['Ship Date']-data['Order Date']).dt.days
region_delay=data.groupby('State')['Shipping Delay'].mean().reset_index()
plt.figure(figsize=(15,10))
plt.barh(region_delay['State'],region_delay['Shipping Delay'])
plt.title('Aversage Shipping Delay by state')
plt.xlabel('Average  Delay(Days)')
plt.ylabel('State')
plt.show()
```

```python
#A relation between state delay and discount 
#each point represents a state
import seaborn as sns
s_d_d=data.groupby('State')[['Shipping Delay','Discount']].mean().reset_index()
plt.figure(figsize=(15,10))
sns.scatterplot(data=s_d_d,x='Shipping Delay',y='Discount')
plt.title('Average Discount vs Shipping Delay by State')
plt.xlabel('Averag Shipping Delay(Days)')
plt.ylabel('Average Discount')
plt.show()
top_discount_states = s_d_d.sort_values(by='Discount', ascending=False).head(10)
top_discount_states
```
